```python
import pandas as pd
```

```python
data = pd.read_csv('training_metrics.csv')
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Group by 'Model Type' and 'Epoch' and calculate the mean of 'Accuracy' and 'Loss'
epoch_performance = data.groupby(['Model Type', 'Epoch']).agg(
    Average_Accuracy=('Accuracy', 'mean'),
    Average_Loss=('Loss', 'mean')
).reset_index()

sns.set(style="whitegrid")

# Plot for Average Accuracy
plt.figure(figsize=(12, 6))
sns.lineplot(data=epoch_performance, x='Epoch', y='Average_Accuracy', hue='Model Type', marker='o')
plt.title('Average Accuracy Across Epochs by Model Type')
plt.xlabel('Epoch')
plt.ylabel('Average Accuracy (%)')
plt.legend(title='Model Type', loc='upper left')

# Plot for Average Loss
plt.figure(figsize=(12, 6))
sns.lineplot(data=epoch_performance, x='Epoch', y='Average_Loss', hue='Model Type', marker='o')
plt.title('Average Loss Across Epochs by Model Type')
plt.xlabel('Epoch')
plt.ylabel('Average Loss')
plt.legend(title='Model Type', loc='upper right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('training_metrics.csv')

# Group the data and calculate the average metrics
grouped_performance = df.groupby(['Model Type', 'Model Size', 'Epoch']).agg(
    Average_Accuracy=('Accuracy', 'mean'),
    Average_Loss=('Loss', 'mean')
).reset_index()

# Visualization: Compare accuracy across epochs for each model size and type
plt.figure(figsize=(14, 7))
sns.lineplot(x='Epoch', y='Average_Accuracy', hue='Model Type', style='Model Size', data=grouped_performance, markers=True, dashes=False)
plt.title('Model Learning Performance Across Epochs (Accuracy)')
plt.xlabel('Epoch')
plt.ylabel('Average Accuracy')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Epoch', y='Average_Loss', hue='Model Type', style='Model Size', data=grouped_performance, markers=True, dashes=False)
plt.title('Model Learning Performance Across Epochs (Loss)')
plt.xlabel('Epoch')
plt.ylabel('Average Loss')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
